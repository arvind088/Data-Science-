```python
url="https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
uClient = uReq(url)

page_html=uClient.read()
page_html
```

```python
pagesoup= soup(page_html,"html.parser")
#pagesoup.h1
pagesoup
```

```python
title=[]
price=[]
rating=[]
for container in containers:
    c=container.div.img["alt"]
    price_container=container.find_all("div",{"class": "_1vC4OE _2rQ-NK"})
    p= price_container[0].text.strip()
    
    rating_container=container.find_all("div",{"class":"hGSR34"})
    r = rating_container[0].text
    title.append([c])
    price.append([p])
    rating.append([r])
print("done !!!")
```

```python
titleflat = [y for x in title for y in x]
priceflat = [y for x in price for y in x]
ratingflat = [y for x in rating for y in x]
cdata =({"title":titleflat,"price":priceflat,"rating":ratingflat})
#df.to_csv()
df = pd.DataFrame.from_dict(cdata, orient='index').transpose()
df
```
