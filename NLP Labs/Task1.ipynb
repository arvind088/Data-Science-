```python
regex = re.compile('\s+')
text = """101 COM    Computers
205 MAT   Mathematics
189 ENG   English""" 
#re.split('\s+', text)
# or 
regex.split(text)
```

```python
# get rid of all extra spaces except newline
regex = re.compile('((?!\n)\s+)')
print(regex.sub(' ', text))
```

```python
#1.
text = """101. COM.    Computers.
205 MAT   Mathematics.
189 ENG   English.""" 
print(re.findall(' ', text))  # Space  ?
print(re.findall('.', text))  # .   Any character except for a new line
#print(re.findall('...', text))
```

```python
#2.
print(re.findall('\.', text))  # matches a period
print(re.findall('[^\.]', text))  # matches anything but a period
```

```python
#3
regtxt = 'Nov 3 (Edited 5:46 PM)'
print(re.findall('\d+', regtxt))
```

```python
#4
regtxt = 'Nov 3 (Edited 5:46 PM)'
print(re.findall('\D+', regtxt))
```

```python
#5
regtxt = 'Nov 3 (Edited 5:46 PM)'
print(re.findall('\w+', regtxt))
```

```python
#6
regtxt = 'Nov 3 (Edited 5:46 PM)'
print(re.findall('\W+', regtxt))
```

```python
#7
regtxt = 'Nov 3 (Edited 5:46 PM)'
print(re.findall('[a-zA-Z]+', regtxt))
```

```python
#8
regtxt = 'Nov 3 (Edited 5:46 PM)'
print(re.findall('\d{1}', regtxt))
print(re.findall('\d{2,5}', regtxt))
```
